# How are In-Demand Skills Trending for Data Analysts?

### Methodology

1. Aggregate skill counts monthly
2. Re-analyze based of percentage of total jobs
3. Plot the monthly skill demand

```python
# Importing Libraries
import ast
import pandas as pd
import seaborn as sns
from datasets import load_dataset
import matplotlib.pyplot as plt  

# Loading Data
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

# Data Cleanup
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
df['job_skills'] = df['job_skills'].apply(lambda x: ast.literal_eval(x) if pd.notna(x) else x)
```

### Filtering data for Data Analysts in the US

```python
df_DA_US = df[(df['job_title_short'] == 'Data Analyst') & (df['job_country'] == 'United States')].copy()
```

```python
df_DA_US['job_posted_month_no'] = df_DA_US['job_posted_date'].dt.month # Adding a month column
```

```python
df_DA_US_explode = df_DA_US.explode('job_skills') # So that every skill is displayed individually
df_DA_US_explode.describe()
```

### Pivoting the data frame so that we get how often skills are posted each month

```python
df_DA_US_pivot = df_DA_US_explode.pivot_table(index='job_posted_month_no', columns='job_skills', aggfunc='size', fill_value=0)

df_DA_US_pivot.loc['Total'] = df_DA_US_pivot.sum()

df_DA_US_pivot = df_DA_US_pivot[df_DA_US_pivot.loc['Total'].sort_values(ascending=False).index] # Ordering by rows from most counts to least

df_DA_US_pivot = df_DA_US_pivot.drop('Total') # Droping Total row since we do not need it anymore

df_DA_US_pivot
```

```python
# Getting total numbers of job postings per month
DA_totals = df_DA_US.groupby('job_posted_month_no').size()
DA_totals
```

### Changing the counts to percentage

```python
# Dividing the total postings of each month by the monthly posting of each skill to get the percentage
df_DA_US_percent = df_DA_US_pivot.div(DA_totals/100, axis=0)

# changes month number to month name
df_DA_US_percent = df_DA_US_percent.reset_index()
df_DA_US_percent['job_posted_month'] = df_DA_US_percent['job_posted_month_no'].apply(lambda x: pd.to_datetime(x, format='%m').strftime('%b'))
df_DA_US_percent = df_DA_US_percent.set_index('job_posted_month')
df_DA_US_percent = df_DA_US_percent.drop(columns='job_posted_month_no')

df_DA_US_percent
```

```python
from matplotlib.ticker import PercentFormatter

# We want all the rows and the first five values
df_plot = df_DA_US_percent.iloc[:,:5]

sns.lineplot(data = df_plot, dashes=False, palette='tab10')
sns.set_theme(style='ticks')
sns.despine()

plt.title('Trending Top Skills for Data Analysts in the US')
plt.ylabel('Likelihood in Job Posting')
plt.xlabel('2023')
plt.legend().remove()
plt.gca().yaxis.set_major_formatter(PercentFormatter(decimals=0))

# Define offsets for the labels to avoid overlapping
offsets = [0, 0.5, -2, 1, -1]

for i in range(5):
    plt.text(11.2, df_plot.iloc[-1, i] + offsets[i], df_plot.columns[i], color='black')

plt.show()
```
